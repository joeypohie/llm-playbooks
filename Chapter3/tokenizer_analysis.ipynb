```python
import requests
import json
```

```python
url = "https://huggingface.co/EleutherAI/gpt-neox-20b/resolve/main/tokenizer.json"
```

```python
response = requests.get(url)
tokenizer_json = response.json()
```

```python
print(tokenizer_json.keys())
```

```python
vocab = tokenizer_json['model']['vocab']
vocab
```

Q3b. The top ten longest tokens represent byte pair encodings

```python
token_length = {}

for key, value in vocab.items():
    token_length[key] = len(key)

print(token_length)

sorted_token_length = dict(sorted(token_length.items(), key=lambda item: item[1], reverse=True))

for token, length in list(sorted_token_length.items())[:10]:
    print(f"'{token}: {length}")
```
